```python
# Import the FixNum class from FixNum.py
from FixNum import FixNum

userinput = input("Number of decimal places (default 2): ")
prec = int(userinput) if userinput else 2 # Convert input to int if or default to 2

def Menu():
    while True:
        print("Menu:\n [a] Fixed-point addition \n [b] Fixed-point power operation \n [x] Exit") # Display a menu and handle user input
        choice = input("Enter your choice: ")
        if choice == 'a':
            print("Fixed-point addition selected (x+y).")
            print("Input 1st fixed point number:")
            x = FixNum(prec) # Initialize X as a FixNum object with the global prec variable
            x.input() # Call the input method to get user input for x
            print(x)
            print("Input 2nd fixed point number:")
            y = FixNum(prec)
            y.input()
            print(y)
            print("x + y =", x + y) 
        elif choice == 'b':
            print("Fixed-point power operation selected (x^y).")
            print("Input 1st fixed point number:")
            x = FixNum(prec)
            x.input()
            print(x)
            print("Input power:")
            y = FixNum(prec)
            y.input()
            print(y)
            # Guard against 0^0 which is undefined in this context
            if float(x) == 0 and float(y) == 0:
                raise Exception("Error: Fixed point number is zero")
            # Convert FixNum instances to float and compute the power, rounded to precision
            print("x ^ y =", round((float(x) ** float(y)), prec))
        elif choice == 'x':
            print("Exiting the menu.")
            break
        else:
            print("Invalid choice. Please try again.")
Menu()
```
