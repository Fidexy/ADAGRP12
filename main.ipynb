```python
from fixNum import FixNum

%load_ext autoreload
%autoreload 2

userinput = input("Number of decimal places (default 2): ")    
prec = int(userinput) if userinput else 2
def Menu():
    while True:
        print("Menu:\n [a] Fixed-point addition \n [b] Fixed-point multiplication \n [x] Exit") # Display a menu and handle user input
        choice = input("Enter your choice: ")
        if choice == 'a':
            print("Fixed-point addition selected (x+y).")
            print("Input 1st fixed point number:")
            x = FixNum(prec)
            x.input()
            print(x)
            print("Input 2nd fixed point number:")
            y = FixNum(prec)
            y.input()
            print(y)
            print("x + y =", x + y)
        elif choice == 'b':
            print("Fixed-point multiplication selected (x^y).")
            print("Input 1st fixed point number:")
            x = FixNum(prec)
            x.input()
            print(x)
            print("Input power y:")
            y = FixNum(prec)
            y.input()
            print(y)
            print("x ^ y =", round((float(x) ** float(y)), prec))
        elif choice == 'x':
            print("Exiting the menu.")
            break
        else:
            print("Invalid choice. Please try again.")

Menu()
```
